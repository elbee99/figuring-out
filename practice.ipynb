```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np

def min_number(lis):
    min_number = lis[0]
    for i in lis:
        for j in lis:
            if i == j:
                continue
            if i < min_number:
                min_number = i
    
    return min_number

#plot of time vs length of list
#
list_lengths = list(range(5,1000,5))
time_taken = []
for i in list_lengths:
    rando = []
    for j in range(i):
        rando.append(random.randint(1,1000))
    start = time.time()
    minimum = min_number(rando)
    end = time.time()
    time_taken.append(end-start)
plt.plot(list_lengths,time_taken)
x = np.linspace(1,1000,1000)
a,b,c = np.polyfit(list_lengths, time_taken,deg=2)
plt.plot(x,a*x**2+b*x+c)
#plt.plot(x,*square(x))
plt.show()
print(list_lengths)
    
    
    

                
            
                
        
```
